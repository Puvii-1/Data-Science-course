```python
seti_a  = {"a","b","c"}
seti_b = {"d","e","f"}
# adding two sets
seti_a.update(seti_b)
seti_a
```

```python
#remove()
seti_a.remove("a")
seti_a
```

```python
seti_a.discard("b")
seti_a
```

```python
# set  operations
set_a = {"Apple","b","c"}
set_b = {"Apple","Microsoft","Google",}
set_a.intersection(set_b)
```

```python
set_z = set_a.union(set_b)
set_z
```
