More on Functions

1. Lambda Function

   It is a function without a name.It is called as anonymous function

```python
# LAMBDA FUNCTION
# syntax->lambda argument(can have multiple arguments) : logic(but can have only single logic)
(lambda a,b : a+b) (4,56)
```

```python
(lambda a,s,d,f,g,h,j,k,l:a+s+d+f+g+h+j+k+l) (1,2,3,4,5,6,7,8,9)
```

```python
lam = lambda a,b : a+b
lam(4,5)
```

```python
def subs(a,b):
  return a-b
```

```python
def addition(a,b):
  if a>b:
    print(a+b)
  else:
    print(a-b)
  return subs(a,b)
```

```python
addition(3,2)
```

Filter Method

```python
List = [1,2,3,4,5,6,7,8,9]
```

```python
def even(List):
  arr = []
  for i in List:
    if i%2==0:
     arr.append(i)
    else:
      pass
  return arr
```

```python
even(List)
```

```python
lambda x: x%2==0
```

```python
list(filter(lambda x: x%2==0 , List))
```

```python
set(filter(lambda x: x%2==0 , List))
```

```python
# ADVANTAGES OF LAMBDA FUNCTION
# 1. simplified code
# 2. Enhamce the readability
# 3. Uses less storage
# 4. Complexity is less
```

```python
# DISADVANTAGES OF LAMBDA
# 1. we cannot reuse the code
# 2. we cannot give multiple logic
# 3. without name
# 4. it doesn't involve docstring
```

```python
# When do we have to use lambda function
# 1. when we know we won't be using the function again
# 2. when the function not so complex and only performing the basic functionality
```
