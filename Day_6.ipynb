```python
# a simple code for checking whether the given input equals to 10
a = int(input())
if a==10:
  print("Something")
else:
  print("Nothing")
```

```python
# Function : using function and calling it is very easy compared to writing the same code for the hundred time
# here you just define a function by giving it name(by your choice) and writing the logic for it
def check(): # defining it function
  a = int(input()) #indented logic
  if a==10:
    print("Something")
  else:
    print("Nothing")
check() # no indentation # calling the function
```

```python
check()
```

```python
# function that takes the parameter
def age(age_value):
  if age_value>60:
    print("OLD")
  elif 30 < age_value<=60:
    print(" MIDDLE AGED ")
  elif 13<age_value<=30:
    print("YOUNG")
  elif 13 <age_value<=19:
    print("TEEN")
  else:
    print("KID")
age(23)
```

```python
age_value = int(input())
age(age_value)
```

```python
advance(1,2,3,4)
```

```python
# *args -> tuple
def advanced(a,b,c,d,*args):
  print(args)
```

```python
advanced(1,2,3,4,6,7,9,7,8,5,4,2,4,6)
```

```python
# **kwargs
# (Dictionary)
def advanced(**kwargs):
  print(kwargs)
```

```python
data={"Fruits":["Mango","Orange","Apple"],
          "Movies":["3 idiots","Marvel","DC"]}
advanced(**data)
```
