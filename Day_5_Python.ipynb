```python
# rock paper scissor
series = int(input("Enter the number of series: "))
a_score = 0
b_score = 0
for i in range(series):
  a = input("Enter your choices player A (r,s,p):")
  b = input("Enter your choices player B (r,s,p):")
  if a == b:
     print("Draw")
  elif a == "r" and b == "p" or a == "p" and b == "s" or  a == "s" and b == "r":
     print("B wins")
     b_score += 1
  elif a == "r" and b == "s" or a == "p" and b == "r" or a == "s" and b == "p":
     print("A wins")
     a_score += 1
  else:
    print("Invalid choice")
print("Winner of the series is: ", end="")
if a_score > b_score:
  print("A")
elif a_score < b_score:
  print("B")
else:
  print("Draw")
```
